# Complexity & Big-O

We are often interested in knowing the worst case runtime of a given algorithm. That's where Big O notation is useful. 

Say, for example, we find that the time it takes for algorithm X to complete a problem of size *n* is $2n^5 + 7n^3 + 4n + 3$. Since we're only interested in the worst case scenario (when n becomes very large), we drop all but the fastest growing term. In this case we only care about $2n^5$. We throw away all the other terms. It's also convention to drop coefficients, so we can say that algorithm X grows at the order of $n^5$. Another way to say this is: O($n^5$). 

It may be useful now to list classes of functions that are commonly encountered when analyzing algorithms. They are listed in order, by the slowest growing to the fastest growing. 'c' represents some constant.

---

|notation|name|
|---|---|
|O(1)|constant|
|O(log(n))|logarithmic|
|O(c*log(n))|polylogarithmic|
|O(n)|linear|
|O(n*log(n)|linearithmic|
|O($n^2$)|quadratic|
|O($n^c$)|polynomial|
|O($c^n$)|exponential|

---

Let's visualize the differences in a plot to make it more concrete.

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(1,10)
plt.figure(figsize=(7,7))
plt.plot(x, [1]*len(x), 'k')
plt.plot(x, np.log(x), 'b')
plt.plot(x, 3*np.log(x), 'c')
plt.plot(x, x, 'm')
plt.plot(x, x*np.log(x), 'y')
plt.plot(x, x**2, 'r')
plt.legend(['constant', 'log(n)', 'c*log(n)', 'linear', 'n*log(n)', 'quadratic'], loc='best', prop={'size':7})
plt.xlim(1,10)
plt.ylim(0,10)
plt.xlabel('n')
plt.ylabel('time')
```

# Performance vs Complexity

It's important to keep in mind the difference between performance and complexity, terms that will be defined shortly. Before doing that, it is important to note that we're coming at this from the perspective of machine learning so the notation and examples will be skewed with that in mind. 

---
**Definitions:**


- **Resources** = {CPU usage (time), memory usage, disk usage, network usage}

- **Performance** = how many resources are used when a program is run (depends on machine, compiler, code, etc)

- **Complexity** = how do the resource requirements scale (i.e. what happens as the number of observations in our dataset increases?)

---

<font color='blue'>Think of performance as **a snapshot in time** whereas complexity is the **change in performance** as the number of observations in the dataset increases. 
<font color='black'>

---

For example, say we wanted to insert an element into an array. The worst case is when inserting at the beginning of the array because all of the elements in the array must be moved. Therefore, in the worst case, the time for insertion is proportional to the number of elements in the array, and we say that the worst-case time for the insertion operation is linear in the number of elements in the array. For a linear-time algorithm, if the problem size doubles, the number of operations also doubles.

---

**Complexity affects performance but not the other way around.**

# Determining Complexity

How do you determine the complexity of your code? Here's a breakdown to help you figure it out.

---

### Sequence of Statements
    statement 1
    statement 2
    ...
    statement n
    
The total time is found by summing the times for all statements:  
    
    total time = time(statement 1) + time(statement 2) + ... + time(statement n)

If each statement only involves basic operations, then the time for each statement is constant and the total time is O(1).

---

### If-Then-Else
    if cond:
        block 1
    else:
        block 2
        
Either block 1 will execute or block 2 will execute. Therefore, the worst-case time is the slower of the two possibilities:  

    max( time(block 1), time(block 2) )  
    
If block 1 takes O(n) and block 2 takes O($n^3$), then the if-then-else statement is O($n^3$).

---

### Loops
    for i in range(n):
        sequence of statements

The loop executes n times. Therefore, the sequence of statements also executes n times. If we assume the statements are O(1), the total time for the for loop is n * O(1), which is O(n).

---

### Nested Loops
    for i in range(n):
        for j in range(m):
            sequence of statements

The outer loop executes n times. Every time the outer loop executes, the inner loop executes m times. As a result, the statements in the inner loop execute a total of n * m times. Thus, the complexity is O(n*m).

---

### Statements with Function Calls

Assume that you know that function f takes constant time and that function g takes linear time. Then the statements below have the time complexities indicated.

    f(n) has O(1)
    g(n) has O(n)

The same rule applies when a loop is involved. For example:
    
    for j in range(n): 
        g(j)

has complexity ($n^2$). The loop executes n times and each function call g(n) has complexity O(N).
